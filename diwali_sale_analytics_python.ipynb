```python
df.columns
```

```python
gender count
```

```python
ax=sn.countplot(x='Gender' , data=df)
    
```

```python
ax=sn.countplot(x='Gender' , data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sn.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
*From above graphs we can see that most of the buyers are females and females are purchasing more over men
```

```python
#Age
```

```python
ax = sn.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sn.set(rc={'figure.figsize':(10,5)})
sn.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

```python
*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
#state
```

```python
df.columns
```

```python
a = sn.countplot( data = df , x='State' , hue='Gender'  )
sn.set(rc={'figure.figsize':(40,5)})
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

sn.set(rc={'figure.figsize':(35,5)})
sn.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sn.set(rc={'figure.figsize':(15,5)})
sn.barplot(data = sales_state, x = 'State',y= 'Amount')
```

```python
*From above graphs we can see that most of the orders & total sales amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*
```

```python
#Marital status
```

```python
m=sn.countplot(x='shadi', data = df , hue='Gender')
sn.set(rc={'figure.figsize':(2,5)})
```

```python
sales_marital_status = df.groupby(['shadi', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sn.set(rc={'figure.figsize':(10,5)})
sn.barplot(data = sales_marital_status, x = 'shadi',y= 'Amount' , hue='Gender')
```

```python
*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*
```

```python
#occupation
```

```python
sn.set(rc={'figure.figsize':(20,5)})
o = sn.countplot(data = df, x = 'Occupation' , hue='Gender')
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sn.set(rc={'figure.figsize':(20,5)})
sn.barplot(data = sales_occupation, x = 'Occupation',y= 'Amount')
```

```python
*From above graphs we can see that most of the buyers are female working in IT, Healthcare and Aviation sector*
```

```python
#product category
```

```python
sn.set(rc={'figure.figsize':(20,5)})
p = sn.countplot(data = df, x = 'Product_Category')
```

```python
sales_product_category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sn.set(rc={'figure.figsize':(20,5)})
sn.barplot(data = sales_product_category, x = 'Product_Category',y= 'Amount')
```

```python
*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*
```
